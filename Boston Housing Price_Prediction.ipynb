```python
#Importing Libraries

import pandas as pd
import numpy as np
from sklearn import metrics
from matplotlib import rcParams
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
rcParams['figure.figsize'] = (20,  20)

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
#Loading the dataset using sklearn library
from sklearn.datasets import load_boston

boston = load_boston()

data = pd.DataFrame(boston.data)
```

```python
# Head of the dataset
display(data)
```

```python
# Allotting Feature names to the dataframe
data.columns = boston.feature_names

#After Adding Feature names to the columns
display(data.head())
```

# **Insights of the Boston Dataset**

```python
#Assigning 'PRICE' as Target variable to the dataframe

data['PRICE'] = boston.target
```

```python
#Extracting shape of the dataset
display("Shape of The Dataset",data.shape)
```

```python
#Extracting Column Names of the dataset
display("Column Names",data.columns)
```

```python
#Extracting Data Types of The dataset
display("Data Types in The Dataset",data.dtypes)
```

```python
#Extracting Unique number of Values in the Dataset
display("Unique Values in The Dataset", data.nunique())
```

```python
#Extracting Missing Values in the dataset
display(data.isnull().sum())
```

```python
#Checking for missing values in the rows of the dataset
display(data[data.isnull().any(axis=1)])
```

```python
#Detail Description of the Dataset
display(data.describe())
```

```python
#Extracting The correlation between the features
correl = data.corr()
display("Correlation Between The Features",correl.shape)
```

**Visualization**

```python
#Heatmap of correlation between the features
plt.figure(figsize=(20,20))
plt.title("Correlation between the Features", fontsize =45)
fig = sns.heatmap(correl, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15})
```

# **Splitting Training & Testing The Data**

```python
# splitting target variable and independept variable
X = data.drop(['PRICE'], axis =1) 
y = data['PRICE']
```

```python
#Splitting into to Training and Testing Data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 4)
```

# **Training the Model (Linear Regression)**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

#Training The model

display(lr.fit(X_train, y_train))
```

```python
# value of y intercept
display("Value of y intercept",lr.intercept_)
```

```python
#Converting the coeff values into a dataframe
coeff = pd.DataFrame([X_train.columns, lr.coef_]).T

coeff = coeff.rename(columns={0: 'Attribute', 1:'Coefficeints'})

display(coeff)
```

```python
# Model prediction on train data
y_pred = lr.predict(X_train)
```

```python
# Model Evaluation
print('R^2:',metrics.r2_score(y_train, y_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_train, y_pred))
print('MSE:',metrics.mean_squared_error(y_train, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

## **Visualization**

```python
# Visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Prices", fontsize =45)
plt.ylabel("Predicted prices", fontsize =45)
plt.title("Prices vs Predicted prices", fontsize =45)
plt.show()
```

```python
# Checking for residuals
plt.scatter(y_pred,y_train-y_pred)
plt.title("Predicted vs residuals", fontsize = 45)
plt.xlabel("Predicted", fontsize = 45)
plt.ylabel("Residuals", fontsize = 45)
plt.show()
```

```python
# Checking Normality errors
sns.distplot(y_train-y_pred)
plt.title("Histogram of Residuals", fontsize = 45)
plt.xlabel("Residuals", fontsize = 45)
plt.ylabel("Frequency", fontsize = 45)
plt.show()
```

# **Prediction Using the Model**

```python
# Predicting Test data with the model
y_test_pred = lr.predict(X_test)
```

```python
# Model Evaluation
acc_linreg = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_linreg)
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```
